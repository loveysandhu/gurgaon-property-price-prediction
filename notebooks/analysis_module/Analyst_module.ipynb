```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('final_cleaned_land.csv')
```

```python
df.head()
```

```python
latlong = pd.read_csv('latlong.csv')
```

```python
latlong['latitude'] = latlong['coordinates'].str.split(',').str.get(0).str.split('°').str.get(0).astype('float')
latlong['longitude'] = latlong['coordinates'].str.split(',').str.get(1).str.split('°').str.get(0).astype('float')
```

```python
new_df_land = df.merge(latlong, on='sector')
```

```python
df1 = pd.read_csv('gurgaon_properties_post_feature_selection_v2.csv')
```

```python
df1.head()
```

```python
new_df_flat = df1.merge(latlong, on='sector')
```

```python
group_df_flat = new_df_flat.groupby('sector').mean(numeric_only=True)[[
    'price',
    'price_per_sqft',
    'built_up_area',
    'latitude',
    'longitude'
]]
```

```python
group_df_land = new_df_land.groupby('sector').mean(numeric_only=True)[[
    'price',
    'price_per_sqm',
    'area',
    'latitude',
    'longitude'
]]
```

```python
group_df_land
```

```python
fig = px.scatter_mapbox(group_df_flat, lat="latitude", lon="longitude", color="price_per_sqft", size='built_up_area',
                  color_continuous_scale=px.colors.cyclical.IceFire, zoom=10,
                  mapbox_style="open-street-map",text=group_df_flat.index)
fig.show()
```

```python
fig = px.scatter_mapbox(group_df_land, lat="latitude", lon="longitude", color="price_per_sqm", size='area',
                  color_continuous_scale=px.colors.cyclical.IceFire, zoom=10,
                  mapbox_style="open-street-map",text=group_df_land.index)
fig.show()
```

```python
df_flat = pd.read_csv('gurgaon_properties.csv')
```

```python
df_land = pd.read_csv('residential_land.csv')
```

```python

wordcloud_df_flat = df_flat.merge(df1, left_index=True, right_index=True)[['features','sector']]
```

```python
wordcloud_df_land = df_land.merge(df, left_index=True, right_index=True)[['features','sector']]
```

```python
import ast
main_flat = []
for item in wordcloud_df_flat['features'].dropna().apply(ast.literal_eval):
    main_flat.extend(item)
```

```python
from wordcloud import WordCloud
```

```python
feature_text_flat = ' '.join(main_flat)
```

```python
import pickle
pickle.dump(feature_text_flat, open('feature_text_flat.pkl','wb'))
```

```python
import ast

main_land = []

for item in wordcloud_df_land['features'].dropna().apply(ast.literal_eval):

    for x in item:
        if isinstance(x, list):
            main_land.extend([i for i in x if str(i).strip() != ''])
        elif str(x).strip() != '':
            main_land.append(x.strip())

```

```python
feature_text_land = ' '.join(main_land)
```

```python
import pickle
pickle.dump(feature_text_land, open('feature_text_land.pkl','wb'))
```
