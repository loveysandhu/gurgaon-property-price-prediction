```python
def extract_area(x):
    area = x.split('sq.m.')[0].replace('(','').strip()

    if '-' in area:
        a, b = area.split('-')
        return (float(a) + float(b)) / 2

    return float(area)
```

```python
df.insert(loc=3,column = 'area',value=df['areaWithType'].apply(extract_area))
```

```python
df['open_sides'] = df['open_sides'].fillna(
    df.groupby('sector')['open_sides'].transform('median')
)
```

```python
df['open_sides'] = df['open_sides'].fillna(df['open_sides'].median())
```

```python
df = df[df['price'].notna()]

df.drop(
    columns=['areaWithType', 'floor_num', 'society', 'possession'],
    inplace=True
)
```

```python
df['features'].value_counts()
```

```python
def feature_score(x):
    if pd.isna(x):
        return 0

    score = 0

    if 'Feng Shui' in x:
        score += 1
    if 'Maintenance Staff' in x:
        score += 1
    if 'Water Storage' in x:
        score += 1
    if 'Rain Water Harvesting' in x:
        score += 1

    return score

df['feature_score'] = df['features'].apply(feature_score)
```

```python
def feature_category(score):
    if score == 0:
        return 'Basic'
    elif score <= 2:
        return 'Standard'
    else:
        return 'Premium'

df['feature'] = df['feature_score'].apply(feature_category)
```
