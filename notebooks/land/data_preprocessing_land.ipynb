```python
df = pd.read_csv('residential_land.csv')
```

```python
df.sample(1)
```

```python
df.drop(columns=['link','address','nearbyLocations','description','rating','property_id'], inplace=True)
```

```python
df['society'].value_counts()
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['price'].value_counts()
```

```python
df = df[~(df['price'] == 'Price on Request')]
```

```python
df.shape
```

```python
def price_convert(x):
  if type(x) == float or type(x) == int:
    return x
  else :
    if x[1] == 'Lac':
      return round(float(x[0])/100,2)
    else :
      return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(price_convert)
```

```python
df['areaWithType'].value_counts()
```

```python
df.rename(columns={'noOfOpenSides':'open_sides','floorNumAllowed':'floor_num'},inplace = True)
```
